What are the least desirable attributes in a male partner? Does this differ for female partners?

```python
attributes_of_partner = ['attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o', 'shar_o']

df_attr = data.loc[:, attributes_of_partner]

df_attr.head()
```

```python
# Mask to keep female population
mask_female = (data['gender'] == 0)
# Dataframe female
df_female = data[mask_female]
df_female_attributes = df_female.loc[:,attributes_of_partner]
df_f = pd.DataFrame({'attributes': attributes_of_partner, 'score': df_female_attributes.sum()})
df_f = df_f.sort_values(by='score', ascending=False).reset_index().drop('index', axis=1)
```

```python
# Mask to keep male population
mask_male = (data['gender'] == 1)
# Dataframe male
df_male = data[mask_male]
df_male_attributes = df_male.loc[:,attributes_of_partner]
df_m = pd.DataFrame({'attributes': attributes_of_partner, 'score': df_male_attributes.sum()})
df_m = df_m.sort_values(by='score', ascending=False).reset_index().drop('index', axis=1)
```

```python
sns.barplot(df_f, x='attributes', y='score')
plt.title('Female rating desires males attributes')
plt.ylabel('score_total_sum')
plt.show()
```

```python
sns.barplot(df_m, x='attributes', y='score')
plt.title('Male rating desires females attributes')
plt.ylabel('score_total_sum')
plt.show()
```

## How important do people think attractiveness is in potential mate selection vs. its real impact?

#### La note attribuée par le participant
- attr_o : La note attribuée par le participant à l'attribut "Attractif" de son partenaire.
- sinc_o : La note attribuée par le participant à l'attribut "Sincère" de son partenaire.
- intel_o : La note attribuée par le participant à l'attribut "Intelligent" de son partenaire.
- fun_o : La note attribuée par le participant à l'attribut "Amusant" de son partenaire.
- amb_o : La note attribuée par le participant à l'attribut "Ambitieux" de son partenaire.
- shar_o : La note attribuée par le participant à l'attribut "A des intérêts/hobbies partagés" de son partenaire.

#### La note attribuée par le La note attribuée par le partenaire
- attr : La note attribuée par le partenaire au participant pour l'attribut "Attractif".
- sinc : La note attribuée par le partenaire au participant pour l'attribut "Sincère".
- intel : La note attribuée par le partenaire au participant pour l'attribut "Intelligent".
- fun : La note attribuée par le partenaire au participant pour l'attribut "Amusant".
- amb : La note attribuée par le partenaire au participant pour l'attribut "Ambitieux".
- shar : La note attribuée par le partenaire au participant pour l'attribut "A des intérêts/hobbies partagés".

```python
col_attr_list = ['attr', 'attr_o', 'dec', 'match']

df_attr_mate_sel = data.loc[:, col_attr_list]

df_attr_mate_sel = df_attr_mate_sel.dropna(subset=col_attr_list)

df_attr_mate_sel['attr'] = df_attr_mate_sel['attr'].astype(int)
df_attr_mate_sel['attr_o'] = df_attr_mate_sel['attr_o'].astype(int)
df_attr_mate_sel['dec'] = df_attr_mate_sel['dec'].astype(int)
df_attr_mate_sel['match'] = df_attr_mate_sel['match'].astype(int)

df_attr_mate_sel.head()
```

```python
sns.kdeplot(df_attr_mate_sel, x='attr', fill=True, common_norm=False, palette="crest", alpha=.5, linewidth=0)
sns.kdeplot(df_attr_mate_sel, x='attr_o', fill=True, common_norm=False, palette="crest", alpha=.5, linewidth=0)
plt.title('Attrativness distribution')
plt.xlabel('Attrativness')
plt.legend(['participant attr', 'partner attr'])
plt.show()
```

```python
sns.displot(df_attr_mate_sel, x='attr', hue='dec', kind='kde')
plt.title('Attrativness of participant / Decision')
plt.xlabel('Attrativness')
plt.show()
```

```python
sns.displot(df_attr_mate_sel, x='attr_o', hue='dec', kind='kde')
plt.title('Attrativness of partner / Decision')
plt.xlabel('Attrativness')
plt.show()
```

```python
sns.displot(df_attr_mate_sel, x='attr', hue='match', kind='kde')
plt.title('Attrativness of participant / Match')
plt.xlabel('Attrativness')
plt.show()
```

```python
sns.displot(df_attr_mate_sel, x='attr_o', hue='match', kind='kde')
plt.title('Attrativness of partner / Match')
plt.xlabel('Attrativness')
plt.show()
```

```python
correlation_attr_mate_sel = df_attr_mate_sel.corr()
```

```python
fig = sns.heatmap(correlation_attr_mate_sel, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matrice de Corrélation')
plt.show()
```

L'attractivité joue un role non negligeable dans la selection d'un partenaire potentiel car la repartition des decision positive se situe pour la plus grosse densité entre 6 et 8, on peut egalement constater que la courbe de decision positive commence à croitre à partir de 4.
afin de pondéré le role de l'attractivité on peu soulever le fait qu'il ne reste qu'une faible proportion de persones pour qui la decision est positive par rapport au nombre de personnes ayant recu une decision negative.
l'attractivité reste cependant une information supplémentaire à ne pas exclure dans le choix d'un partenaire, mais ne permet pas d'expliquer de maniere certaines la prise de decision .

Are shared interests more important than a shared racial background?

```python
# imprace race samerace shar1_1 (during date) shar1_2 (before date)
col_shar_list = ['race', 'imprace', 'samerace', 'shar', 'shar_o', 'dec', 'match']
df_share = data.loc[:, col_shar_list]
df_share = df_share.dropna(subset=col_shar_list)
```

```python
fig = sns.heatmap(df_share.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matrice de Corrélation')
plt.show()
```

```python
# DMapping des valeursss
mapping = {
    1: 'Black/African American',
    2: 'European/Caucasian-American',
    3: 'Latino/Hispanic American',
    4: 'Asian/Pacific Islander/Asian-American',
    5: 'Native American',
    6: 'Other'
}

# Nouvelle colonne mappée en utilisant le mapping pour les etnhies
df_share['mapped_ethnicity'] = df['race'].map(mapping)
df_share.head()
```

```python
plot = sns.displot(df_share, x='mapped_ethnicity')
plt.subplots_adjust(hspace = 0.2)
plt.title('Number of people / Ethnicity')
# Axe x à 90 degrés pour afficher les ethnies visiblement
plot.set_axis_labels(x_var="Ethnicity")
plot.set_xticklabels(rotation=90)
plt.show()
```

```python
plot = sns.displot(df_share, x='imprace', col='mapped_ethnicity', col_wrap=3, kde='True')
plt.subplots_adjust(top = 0.9)
plt.suptitle('Importance of ethnicity / Ethnicity')
plt.show()
```

La proportion de l'importance de l'ethnie par ethnie est similaire de facon generale, le plus gros de la densité se situe sur la gauche, la ou les notes restes relativement faible pour la note de cet attribut

```python
plot = sns.displot(df_share, x='imprace', col='dec', col_wrap=2, kde='True')
plt.subplots_adjust(top=0.9)
plt.suptitle('Importance of ethnicity / Decision')
plt.show()
```

L'importance de l'ethnie sur la prise de décision positive ou negative reste également similaire

```python
plot = sns.displot(df_share, x='imprace', col='match', col_wrap=2, kde='True')
plt.subplots_adjust(top=0.9)
plt.suptitle('Importance of ethnicity / Match')
plt.show()
```

L'importance de l'ethnie par rapport au nombre match montre que la proportion de match positif est tres faible par rapport à la proportion de match negatif. on peut en conclure que tout comme l'attractivite, l'importance de l'ethnie reste une information utile mais trop abstraite.

#### In terms of getting a second date, is it better to be someone's first speed date of the night or their last?
* Columns: ['attr', 'sinc', 'intel', 'fun', 'amb', 'shar', 'attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o', 'shar_o', 'order', 'dec', 'match']

```python
attributes_first_last = ['order', 'attr', 'sinc', 'intel', 'fun', 'amb', 'shar', 'attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o', 'shar_o', 'dec', 'match']

# Waves
# October 16th ‘02
mask_wave_1 = (data['wave'] == 1)
# October 23rd ‘02
mask_wave_2 = (data['wave'] == 2)
# November 12th ‘02
mask_wave_3 = (data['wave'] == 3)
```

```python
# Analyze wave 1
df_wave_1 = data[mask_wave_1]
df_wave_1 = df_wave_1.loc[:, attributes_first_last]
df_wave_1 = df_wave_1.dropna(subset=attributes_first_last)
# Trouver l'indice de la ligne ayant la valeur minimale de "order"
min_order_index_w1 = df_wave_1['order'].idxmin()
max_order_index_w1 = df_wave_1['order'].idxmax()

# Obtenir la valeur minimale de "order" en utilisant l'indice
min_order_value = df_wave_1.loc[min_order_index_w1, 'order']

# # Utiliser cette valeur pour effectuer un group by
w1_grouped_first_date = df_wave_1[df_wave_1['order'] == min_order_value]

max_order_index_w1 = df_wave_1['order'].idxmax()

max_order_value = df_wave_1.loc[max_order_index_w1, 'order']

w1_grouped_last_date = df_wave_1[df_wave_1['order'] == max_order_value]
```

```python
w1_grouped_first_date.head()
```

```python
w1_grouped_last_date.head()
```

```python
# Concat df
w1_grouped = pd.concat([w1_grouped_first_date, w1_grouped_last_date])
```

```python
# w1_grouped.head()
attributes_list = ['attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o', 'shar_o', 'dec', 'match']
mask_first = (w1_grouped['order'] == w1_grouped['order'].min())
w1_first_date_filtered = w1_grouped[mask_first]
w1_first_date = w1_first_date_filtered.loc[:, attributes_list]
```
