What are the least desirable attributes in a male partner? Does this differ for female partners?

```python
attributes_of_partner = ['attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o', 'shar_o']

df_attr = data.loc[:, attributes_of_partner]

df_attr.head()
```

```python
# Mask to keep female population
mask_female = (data['gender'] == 0)
# Dataframe female
df_female = data[mask_female]
df_female_attributes = df_female.loc[:,attributes_of_partner]
df_f = pd.DataFrame({'attributes': attributes_of_partner, 'score': df_female_attributes.sum()})
df_f = df_f.sort_values(by='score', ascending=False).reset_index().drop('index', axis=1)
```

```python
# Mask to keep male population
mask_male = (data['gender'] == 1)
# Dataframe male
df_male = data[mask_male]
df_male_attributes = df_male.loc[:,attributes_of_partner]
df_m = pd.DataFrame({'attributes': attributes_of_partner, 'score': df_male_attributes.sum()})
df_m = df_m.sort_values(by='score', ascending=False).reset_index().drop('index', axis=1)
```

```python
sns.barplot(df_f, x='attributes', y='score')
plt.title('Female rating desires males attributes')
plt.ylabel('score_total_sum')
plt.show()
```

```python
sns.barplot(df_m, x='attributes', y='score')
plt.title('Male rating desires females attributes')
plt.ylabel('score_total_sum')
plt.show()
```

```python
# DMapping des valeursss
mapping = {
    1: 'Black/African American',
    2: 'European/Caucasian-American',
    3: 'Latino/Hispanic American',
    4: 'Asian/Pacific Islander/Asian-American',
    5: 'Native American',
    6: 'Other'
}

# Nouvelle colonne mappÃ©e en utilisant le mapping pour les etnhies
df_share['mapped_ethnicity'] = df['race'].map(mapping)
df_share.head()
```

#### In terms of getting a second date, is it better to be someone's first speed date of the night or their last?
* Columns: ['attr', 'sinc', 'intel', 'fun', 'amb', 'shar', 'attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o', 'shar_o', 'order', 'dec', 'match']

```python
attributes_first_last = ['order', 'attr', 'sinc', 'intel', 'fun', 'amb', 'shar', 'attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o', 'shar_o', 'dec', 'match']

# Waves
# October 16th â€˜02
mask_wave_1 = (data['wave'] == 1)
# October 23rd â€˜02
mask_wave_2 = (data['wave'] == 2)
# November 12th â€˜02
mask_wave_3 = (data['wave'] == 3)
```

```python
# Analyze wave 1
df_wave_1 = data[mask_wave_1]
df_wave_1 = df_wave_1.loc[:, attributes_first_last]
df_wave_1 = df_wave_1.dropna(subset=attributes_first_last)
# Trouver l'indice de la ligne ayant la valeur minimale de "order"
min_order_index_w1 = df_wave_1['order'].idxmin()
max_order_index_w1 = df_wave_1['order'].idxmax()

# Obtenir la valeur minimale de "order" en utilisant l'indice
min_order_value = df_wave_1.loc[min_order_index_w1, 'order']

# # Utiliser cette valeur pour effectuer un group by
w1_grouped_first_date = df_wave_1[df_wave_1['order'] == min_order_value]

max_order_index_w1 = df_wave_1['order'].idxmax()

max_order_value = df_wave_1.loc[max_order_index_w1, 'order']

w1_grouped_last_date = df_wave_1[df_wave_1['order'] == max_order_value]
```
